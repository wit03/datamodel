### 3. Inspect if there are any missing values.

```python
# 3.1 Remove rows that contain missing value
row_remove = df.dropna()
print("Shape: ", row_remove.shape)
row_remove.info()
```

```python
# 3.2 Replace missing value with zeros
fill_zero = df.fillna(0)
print("Shape: ", fill_zero.shape)
fill_zero.describe()
```

```python
# Calculate the mean of every attribute in fill_zero and df
fill_zero_mean = fill_zero.select_dtypes(include=np.number).mean()
df_mean = df.select_dtypes(include=np.number).mean()

mean_comparison = pd.DataFrame({'fill_zero': fill_zero_mean, 'ori_df': df_mean})
print(mean_comparison)
```

<b>*ANS*</b>: Filling Zero shoud not be use, since some data such as Average Building Area is largely differs than the original data. Thus, the meaning of average is not truly indicates the real value of it.

```python
# 3.3 Replace missing value on non-numeric column with zeros, and on numeric column with mean.
data_imp = df

for column in data_imp.columns:
    if data_imp[column].dtype == np.number:
        data_imp[column] = data_imp[column].fillna(data_imp[column].mean())
    else:
        data_imp[column] = data_imp[column].fillna(0)

data_imp_mean = data_imp.select_dtypes(include=np.number).mean()
df_mean = df.select_dtypes(include=np.number).mean()

mean_comparison2 = pd.DataFrame({'data_imp': data_imp_mean, 'ori_df': df_mean})
print(mean_comparison2)
```

```python
data_imp.info()
```

<b>*ANS*</b>: Data Imputation shoud be use, since every mean of data stay the same from the original data.
